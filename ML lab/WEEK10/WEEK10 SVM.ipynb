# SVM

1. Use the IRIS dataset, implement the SVM classifier in python (make use of scikit-learn library), to do thefollowing.
a. Apply the kernel functions such as linear, polynomial, Radial basis functions and Sigmoid.
b. Plot the scatter plot of the input features.
c. Plot the decision boundar.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  # Take the first two features for simplicity
y = iris.target

def plot_decision_boundary(clf, X, y, title):
    h = 0.02  # Step size in the mesh

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.turbo, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.turbo)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
```

```python
# Linear kernel
clf_linear = svm.SVC(kernel='linear', C=1)
clf_linear.fit(X, y)

# Polynomial kernel
clf_poly = svm.SVC(kernel='poly', degree=2, C=1)
clf_poly.fit(X, y)

# Radial basis function (RBF) kernel
clf_rbf = svm.SVC(kernel='rbf', gamma=0.5, C=1)
clf_rbf.fit(X, y)

# Sigmoid kernel
clf_sigmoid = svm.SVC(kernel='sigmoid', C=1)
clf_sigmoid.fit(X, y)

# Plot decision boundaries for each kernel
plot_decision_boundary(clf_linear, X, y, 'Linear Kernel')
plt.show()
```

```python
plot_decision_boundary(clf_poly, X, y, 'Polynomial Kernel')
```

```python
plot_decision_boundary(clf_rbf, X, y, 'RBF Kernel')
```

```python
plot_decision_boundary(clf_sigmoid, X, y, 'Sigmoid Kernel')
```
